## SPARQL endpoints


There are several ways to query a SPARQL endpoint, in this example we will use SPARQLWrapper to get results using python...



We define the endpoint we want to query, in this case is the [endpoint of the Publications Office](https://publications.europa.eu/webapi/rdf/sparql)

There are several ways to query a SPARQL endpoint, in this example we will use SPARQLWrapper to get results using python.

````python
# Import libraries
from SPARQLWrapper import SPARQLWrapper, JSON

# Define the SPARQL endpoint URL
endpoint_url = "https://publications.europa.eu/webapi/rdf/sparql"

# Write the SPARQL query
sparql_query = """
PREFIX epo: <http://data.europa.eu/a4g/ontology#>

SELECT DISTINCT ?Notice WHERE {
    ?Notice epo:hasPublicationDate  "20230911".
} LIMIT 10
"""
# Execute the SPARQL query
sparql = SPARQLWrapper(endpoint_url)
sparql.setQuery(sparql_query)
sparql.setReturnFormat(JSON)
results = sparql.query().convert()

# Process and display the results
for result in results["results"]["bindings"]:
    print(result["Notice"]["value"])
````
